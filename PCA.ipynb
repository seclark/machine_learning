Things the curse of dimensionality impacts:

* The size of data required to constrain a model 
 
* The complexity of the model
    
* (The search times required to optimize the model)

As an illustration, we ask: under a Euclidean distance metric, what proportion of points falls within a unit distance of the origin?


The ratio between (1) a hypersphere with radius r and (2) a hypercube with sides of length 2r, in D dimensions:

$f_2\approx78.5\%$

$f_3\approx52.3\%$

$f_D=\frac{V_D(r)}{2r^D}=\frac{\pi^{D/2}}{D2^{D-1}\Gamma(D/2)}$

$\lim_{D \to \inf} f_D = 0$

E.g. for a sample of $3.57\times10^{8}$ SDSS spectra, each described by 30 attributes (flux, size, ellipticity, etc.), the probability of having one source reside within the hypersphere would be only $7\times10^{-6}$.

# 7.3 Principal Component Analysis (PCA)

<strong>PCA Steps</strong>: First, find the first axis that maximizes the variance of the data. Then, find a second axis that is orthogonal to the first axis and maximizes the residual variance, and so on. 


<img src="http://www.astroml.org/_images/fig_PCA_rotation_1.png" style="width: 500.0px; height: 500.0px;">
<p>Figure 7.2</p>
<p>A distribution of points drawn from a bivariate Gaussian and centered on the
origin of x and y. PCA defines a rotation such that the new axes (x&#8217; and y&#8217;)
are aligned along the directions of maximal variance (the principal components)
with zero covariance. This is equivalent to minimizing the square of the
perpendicular distances between the points and the principal components.</p>


##The eigenvector of the covariance matrix with the largest eigenvalue is the direction along which the data has its maximum variance.##


**3 ways to do PCA**

+ SVD of $X$
+ eigenvalue decomposition of $C_X$
+ eigenvalue decomposition of $M_X$

$X$ is a $N \times K$ matrix, of $N$ observations and $K$ features (dimensions). 

$X$ can be decomposed to:

$U\Sigma V^T = \frac {1} {\sqrt{N-1}}X$, where $V^T$ and $U$ are rotations, and $\Sigma$ is a scaling.

$C_X = \frac{1}{N-1}X^TX = R C_Y R^T = R \Sigma^2 R$ (covariance matrix)

$M_X = \frac{1}{N-1}XX^T = U \Sigma^2 U^T$ (correlation matrix)


where R are the eigenvectors (or principal components), diagonal terms of $C_Y$ define the amount of variance contained within each component.

$R=V=\frac{1}{\sqrt{N-1}} X^T U \Sigma^{-1}$

## Application to SDSS data

```python
import numpy as np
import scipy as scipy
from matplotlib import pyplot as plt

from sklearn.decomposition import PCA
from astroML.datasets import sdss_corrected_spectra
from astroML.plotting.tools import discretize_cmap

data = sdss_corrected_spectra.fetch_sdss_corrected_spectra()
spectra = sdss_corrected_spectra.reconstruct_spectra(data)
lam = sdss_corrected_spectra.compute_wavelengths(data)

n_components=20

#instantiate PCA class
pca = PCA(n_components=n_components)

#fit eigenspectra
pca.fit(spectra)

#project to return eigenvalues for each real spectra
eigs = pca.transform(spectra)
eigs.shape

#reconstructed spectrum is Sumerian(eigenspectra*eignevalue)
#specnum = 1
#recon = pca.mean_
#for i in np.arange(n_components):
#	recon = recon + eigs[specnum,i]*pca.components_[i]

#plt.plot(lam, spectra[specnum], c='k')
#plt.plot(lam, recon)
#and now you've reduced your data size by ~50x

def rfilter(spectra):
	rmag = np.sum(spectra[(lam>5500) & (lam<6750)])*.48
	return rmag

def bfilter(spectra):
	bmag = np.sum(spectra[(lam>3200) & (lam<3800)])*.1
	return bmag

bmags = np.zeros(len(spectra))
rmags = np.zeros(len(spectra))
for i in np.arange(len(spectra)):
	bmags[i] = bfilter(spectra[i])
	rmags[i] = rfilter(spectra[i])


plt.figure(figsize=(10,10))
plt.subplot(6,1,1)
plt.plot(lam, pca.mean_, color='k')
plt.locator_params(nbins=4)
plt.xlabel('$\lambda$')
plt.xlim([lam[0]-100, lam[-1]+100])
for i in np.arange(5):
	plt.subplot(6,1,i+2)
	plt.plot(lam, pca.components_[i], color='k')
	plt.locator_params(nbins=4)
	plt.xlim([lam[0]-100, lam[-1]+100])
	plt.xlabel('$\lambda$')

```

```python
#####plot eigenvalues
fig = plt.figure(figsize=(10, 7.5))
cax = plt.axes([0.525, 0.525, 0.02, 0.35])
clim = (1.5, 6.5)
cmap = discretize_cmap(plt.cm.jet, 5)
cdict = ['unknown', 'star', 'absorption galaxy',
         'galaxy', 'emission galaxy',
         'narrow-line QSO', 'broad-line QSO']
cticks = [2, 3, 4, 5, 6]
formatter = plt.FuncFormatter(lambda t, *args: cdict[int(np.round(t))])
plt.subplot(2,2,1)
plt.xlabel('$\mathrm{c_0}$')
plt.ylabel('$\mathrm{c_1}$')
plt.xlim([-500,1500])
plt.ylim([-200,200])
im1 = plt.scatter(eigs[:,0],eigs[:,1], edgecolor = 'none', alpha = 0.4, c=data['lineindex_cln'], cmap = cmap)
im1.set_clim(clim)
plt.subplot(2,2,3)
plt.xlabel('$\mathrm{c_0}$')
plt.ylabel('$\mathrm{c_2}$')
plt.xlim([-500,1500])
plt.ylim([-300,200])
im2 = plt.scatter(eigs[:,0],eigs[:,2], edgecolor = 'none', alpha = 0.4, c=data['lineindex_cln'], cmap = cmap)
im2.set_clim(clim)
ax3 = plt.subplot(2,2,4)
plt.xlabel('$\mathrm{c_1}$')
plt.ylabel('$\mathrm{c_2}$')
plt.xlim([-300,300])
plt.ylim([-300,200])
im3 = plt.scatter(eigs[:,1],eigs[:,2], edgecolor = 'none', alpha = 0.4, c=data['lineindex_cln'], cmap = cmap)
im3.set_clim(clim)
fig.colorbar(im3, ax=ax3, cax=cax,
             ticks=cticks,
             format=formatter)

```

```python
#########################
#########################
# load data
import numpy as np
from sklearn.decomposition import PCA

a = np.loadtxt('./ledaoutput.txt', dtype='string', delimiter='|', skiprows=3)
good  = np.arange(len(a[:,0]))[(a[:,1] != '                 ') &  (a[:,2] != '                 ') & (a[:,3] != '                 ')]

vdisp = a[good,1].astype('float')
radius = a[good,2].astype('float')
absm = a[good,3].astype('float')

#rescale, center at 0
vdispr  = vdisp-np.mean(vdisp)
vdispr  = vdispr/np.var(vdispr)
radiusr = radius-np.mean(radius)
radiusr = radiusr/np.var(radiusr)
absmr   = absm-np.mean(absm)
absmr   = absmr/np.var(absmr)

#compute PCA components
matrix = np.vstack([vdispr, radiusr, absmr])
pca = PCA(n_components=3)
pca.fit(np.transpose(matrix))
comp = pca.components_

new1 = np.dot(np.transpose(matrix), comp[0])
new2 = np.dot(np.transpose(matrix), comp[1])
new3 = np.dot(np.transpose(matrix), comp[2])


###############################plot

plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
plt.scatter(vdisp, radius, color = 'teal', alpha = 0.6)
#plt.xlabel('Velocity dispersion [km/s]')
plt.ylabel('Apparent radius [degrees]')
plt.ylim([21,27])
plt.locator_params(nbins=4)
plt.subplot(2,2,3)
plt.scatter(vdisp, absm, color = 'orange', alpha = 0.6)
plt.xlabel('Velocity dispersion [km/s]')
plt.ylabel('Absolute magnitude [mag]')
plt.locator_params(nbins=4)
plt.subplot(2,2,4)
plt.scatter(radius,absm, color = 'purple', alpha = 0.6)
plt.xlabel('Apparent radius [degrees]')
#plt.ylabel('Absolute magnitude [mag]')
plt.locator_params(nbins=4)
```

**Choosing the Level of Truncation**

+ Too many components will introduce noise

+ Too few may not capture the complete physical correlations

+ Most common criterion: set a bound (normally 0.7-0.95) on the amount of variance we want to capture

+ Cut at the turning point in the scree plot

+ If not sharp turn, then use a log scree plot instead (if noise decay geometrically, variation in the eigenvalues should drop as a linear function??)
